## <center><H1>Shopping Bill Management System</H1></center>

**Objective**

Develop a Python program that calculates a customer's final shopping bill by considering the product price, quantity, membership discount, festival discount, and GST. The program should generate a detailed shopping bill in a well-formatted output.

**Problem Statement**

Create a Python program that accepts customer details, including the customer's name, customer ID, product price, quantity, and membership type. The program should calculate the total purchase amount, apply membership discounts based on the customer's membership category, provide an additional festival discount for eligible purchases, calculate GST, and display the final shopping bill with all calculations in a structured format.


# **Write a Python program that takes the following inputs:** <br>

• customer_name (string) <br>
• customer_id (int) <br>
• product_price (float) <br>
• quantity (int) <br>
• membership_type (Gold / Silver / Regular) <br>

## **Then calculate the final shopping bill using the rules below:** <br>

### **1. Total Purchase Amount:** <br>

• Total Amount = Product Price × Quantity <br>

---

### **2. Membership Discount:** <br>

• Gold Member: **20% Discount** <br>
• Silver Member: **10% Discount** <br>
• Regular Member: **No Discount** <br>

---

### **3. Festival Offer:** <br>

• If Total Amount is more than **₹20,000**, provide an additional **5% discount**. <br>

---

### **4. GST:** <br>

• Apply **18% GST** after all discounts. <br>

---

### **5. Final Bill Calculation:** <br>

• Discounted Amount = Total Amount − Membership Discount − Festival Discount <br>

• GST = Discounted Amount × 18% <br>

• Final Bill = Discounted Amount + GST <br>

---

### **6. Final Output:** <br>

Print a complete Shopping Bill showing: <br>

• Customer Name <br>
• Customer ID <br>
• Product Price <br>
• Quantity <br>
• Total Amount <br>
• Membership Discount <br>
• Festival Discount <br>
• GST <br>
• Final Bill <br>

---

GST                  :<br>
Final Bill           :<br>

```python
customer_name = input("Enter Your Customer Name..")
customer_id = int(input("Enter Your Customer ID "))
product_price = float(input("Enter Your Product Price"))
quantity = int(input("Enter Your Quantity"))
membership_type = input("Enter your Membership Types (Gold / Silver / Regular)")

total_amount = 0
membership_discount= 0
festival_discount = 0
gst = 0



#Total Purchase Amount

total_amount = product_price * quantity

#Membership Discount:

if membership_type == "Gold":
    membership_discount = total_amount * 0.20
elif membership_type == "Silver":
    membership_discount = total_amount * 0.10
else:
    membership_discount = 0

#Festival Discount:

if total_amount > 20000:
    festival_discount = total_amount * 0.05
else:
    festival_discount = 0

# Discounted Amount 

discounted_amount = total_amount - membership_discount - festival_discount

#gst 

gst = discounted_amount*0.18

#Total bill

total_bill = discounted_amount + gst



print("\n=======Total Shopping Bill=======")
print("Customer Name                 :", customer_name)
print("Customer ID                   :", customer_id)
print("Product Price                 :", product_price)
print("Quantity                      :", quantity)
print("Total Amount                  :", total_amount)
print("Membership Discount           :", membership_discount)
print("Festival Discount             :", festival_discount)
print("GST                           :", gst)
print("Total Shopping Bill           :", total_bill)
```
